Working with Missing Data

```python
missing_cs_data = pd.read_csv("car-sales-missing-data.csv")
```

```python
missing_cs_data
```

we use .fillna() to fill in the missing columns "NaN"

Let us use the mean value of the Odometer column to fill in the rest of the missing rows of the same column

Note that this is not a practical solution and its only used for demonstration

```python
missing_cs_data["Odometer"].fillna(missing_cs_data["Odometer"].mean())
```

```python
missing_cs_data
```

The missing column values did not change because we did not assign the value to the columns. But there is also a different way to do this with the use of inplace() although best to avoid its usuage

```python
missing_cs_data["Odometer"] =  missing_cs_data["Odometer"].fillna(missing_cs_data["Odometer"].mean())
```

```python
missing_cs_data
```

```python
missing_dropped = missing_cs_data.dropna()
```

```python
missing_dropped
```

```python
missing_dropped.to_csv("missing-data-dropped.csv", index=False)
```
